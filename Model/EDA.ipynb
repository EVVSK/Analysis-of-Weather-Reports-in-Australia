```python
df.hist(bins = 50, figsize = (20,20))
plt.show()
```

```python
num_cols = len(df.columns)
num_rows = (num_cols // 3)

fig_width = num_cols * 5 
fig_height = num_rows * 15

fig, axes = plt.subplots(nrows=num_rows, ncols=3, figsize=(fig_width, fig_height))

axes = axes.flatten()

# Plot histograms for each column
for i, column in enumerate(df.columns):
    axes[i].hist(df[column], bins=20, edgecolor='black')
    axes[i].set_title(f'Histogram of {column}')
    axes[i].set_xlabel('Values')
    axes[i].set_ylabel('Frequency')
```

```python
corr = df.corr()

# Create a heatmap
plt.figure(figsize=(12, 8))  # Adjust figure size for better readability
sns.heatmap(
    corr, 
    annot=True,  # Add values inside the cells
    fmt=".2f",   # Format the numbers to 2 decimal places
    cmap='coolwarm',  # Use the 'coolwarm' color palette
    cbar_kws={'label': 'Correlation Coefficient'},  # Add label to color bar
    square=True   # Make the cells square
)

# Add title and adjust layout
plt.title('Correlation Heatmap', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Key Findings and Insights
```

```python
temp_conditions_9am = df['Temp9am']
temp_conditions_3pm = df['Temp3pm']
```

```python
print("Summary Statistics for Temperature Conditions at 9 AM:")
print(temp_conditions_9am.describe())
print("\nSummary Statistics for Temperature Conditions at 3 PM:")
print(temp_conditions_3pm.describe())
```

```python
plt.subplot(2, 1, 2)
plt.hist(temp_conditions_9am.dropna(), bins=30, edgecolor='black')
plt.title('Temperature Distribution at 9 AM')
plt.xlabel('Temperature (C)')
plt.ylabel('Frequency')
```

```python
plt.subplot(2, 1, 2)
plt.hist(temp_conditions_3pm.dropna(), bins=30, edgecolor='black')
plt.title('Temperature Distribution at 3 PM')
plt.xlabel('Temperature (C)')
plt.ylabel('Frequency')
```

```python
plt.scatter(df['Temp9am'], df['Rainfall'])
plt.title('Rainfall vs Temperature of same day')
plt.xlabel('Temperature at 9 AM (C)')
plt.ylabel('Rainfall (mm)')
plt.show()
```

```python
plt.scatter(df['Temp3pm'], df['Rainfall'])
plt.title('Rainfall vs Temperature of same day')
plt.xlabel('Temperature at 3 PM (C)')
plt.ylabel('Rainfall (mm)')
plt.show()
```

```python
mean_temp_9am = temp_conditions_9am.mean()
mean_temp_3pm = temp_conditions_3pm.mean()

print("\nInsights:")
print(f"The average temperature at 9 AM is {mean_temp_9am:.2f} C.")
print(f"The average temperature at 3 PM is {mean_temp_3pm:.2f} C.")
```

```python
# Formulating Hypotheses
```

```python
# 1. Temperature has a significant impact on the likelihood of rainfall.
# 2. Humidity levels are predictive of rainfall.
# 3. Sunshine, Pressure and Evaporation play a significant role in affecting rainfall.
# 4. Wind speed is not a significant factor in predicting rainfall.
```

```python
# Conducting a Formal Significance Test
```

```python
t_stat, p_value = stats.ttest_ind(temp_conditions_9am.dropna(), temp_conditions_3pm.dropna())
alpha = 0.05

print("\nHypothesis Testing:")
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject null hypothesis: There is a significant difference in temperatures.")
else:
    print("Fail to reject null hypothesis: There is no significant difference in temperatures.")
```
